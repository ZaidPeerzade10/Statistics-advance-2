# Statistic advance 2

1. What is hypothesis testing in statistics
   - Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data.
   - It involves stating a null and alternative hypothesis and using statistical tests to determine which is more likely.
   - The process helps in assessing whether observed data significantly deviates from what was expected under the null hypothesis.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis
   - The null hypothesis (H‚ÇÄ) is a default assumption that there is no effect or no difference in the population.
   - The alternative hypothesis (H‚ÇÅ or Ha) contradicts the null, suggesting there is an effect or a difference.
   - Hypothesis testing determines whether the sample data provides enough evidence to reject the null in favor of the alternative.

3. What is the significance level in hypothesis testing, and why is it important
   - The significance level (denoted as Œ±) is the probability of rejecting the null hypothesis when it is actually true (Type I error).
   - Common values are 0.05 or 0.01, and it defines the threshold for how extreme the data must be to consider the result statistically significant.
   - It's important because it controls the likelihood of making a false positive conclusion.

4. What does a P-value represent in hypothesis testing
   - A P-value is the probability of obtaining results as extreme as, or more extreme than, those observed in the sample data, assuming the null hypothesis is true.
   - It quantifies the evidence against the null hypothesis.
   - A small P-value indicates strong evidence against H‚ÇÄ.

5. How do you interpret the P-value in hypothesis testing
   - If the P-value is less than or equal to the significance level (Œ±), we reject the null hypothesis, suggesting the result is statistically significant.
   - If it's greater than Œ±, we fail to reject the null, meaning there isn't enough evidence to support the alternative.
   - It doesn't prove a hypothesis but indicates strength of evidence.

6. What are Type 1 and Type 2 errors in hypothesis testing
   - A Type 1 error occurs when the null hypothesis is true but is incorrectly rejected.
   - A Type 2 error happens when the null hypothesis is false but is incorrectly not rejected.
   - Type 1 is a false positive, and Type 2 is a false negative.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing
   - A one-tailed test checks for an effect in only one direction (either greater than or less than a value).
   - A two-tailed test looks for an effect in both directions, checking for any significant difference regardless of direction.
   - The choice depends on the research question and hypothesis.

8. What is the Z-test, and when is it used in hypothesis testing
   - The Z-test is used to determine whether two population means are different when the standard deviations are known and the sample size is large (typically n > 30).
   - It assumes that the data follows a normal distribution.
   - It's often used for comparing sample and population means.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing
   - The Z-score is calculated as Z=ùëã‚àíùúá/œÉ, where X is the sample value, Œº is the population mean, and œÉ is the standard deviation.
   - It represents how many standard deviations a value is from the mean.
   - In hypothesis testing, it helps determine how extreme a value is under the null hypothesis.

10. What is the T-distribution, and when should it be used instead of the normal distribution
   - The T-distribution is similar to the normal distribution but has heavier tails, making it more appropriate for small sample sizes or unknown population standard deviation.
   - As sample size increases, it approaches the normal distribution.
   - It's used in T-tests and other scenarios where Z-tests are not suitable.

11. What is the difference between a Z-test and a T-test
   - A Z-test is used when the population standard deviation is known and the sample size is large.
   - A T-test is used when the population standard deviation is unknown and the sample size is small.
   - Both tests assess whether sample means differ significantly from population means or from each other.

12. What is the T-test, and how is it used in hypothesis testing
   - A T-test compares the means of two groups to determine if they are statistically different from each other.
   - It's used when the population standard deviation is unknown and the sample size is small.
   - There are different types, such as one-sample, two-sample, and paired T-tests, depending on the study design.

13. What is the relationship between Z-test and T-test in hypothesis testing
   - Both Z-tests and T-tests are used to test hypotheses about means, but they differ in their assumptions.
   - Z-tests assume known population standard deviation and large sample size, while T-tests are used for smaller samples and unknown standard deviation.
   - As sample size increases, results from both tests converge.

14. What is a confidence interval, and how is it used to interpret statistical results
   - A confidence interval provides a range of values within which the true population parameter is likely to fall, based on sample data.
   - It gives an estimate with a specific level of confidence (e.g., 95%).
   - A wider interval indicates more uncertainty, while a narrower one shows more precision.

15. What is the margin of error, and how does it affect the confidence interval
   - The margin of error is the amount added and subtracted from the sample estimate to create a confidence interval.
   - It reflects the uncertainty in the estimate due to sampling variability.
   - A larger margin of error results in a wider confidence interval, indicating less precision.

16. How is Bayes' Theorem used in statistics, and what is its significance
   - Bayes' Theorem updates the probability of a hypothesis based on new evidence.
   - It combines prior probability and likelihood to compute a posterior probability.
   - It's especially useful in fields like medical testing, spam filtering, and machine learning for probabilistic inference.

17. What is the Chi-square distribution, and when is it used
   - The Chi-square distribution is used for categorical data to test relationships between variables or the goodness of fit.
   - It's skewed right and depends on degrees of freedom.
   - It's commonly used in Chi-square tests for independence and goodness-of-fit tests.

18. What is the Chi-square goodness of fit test, and how is it applied
   - The Chi-square goodness of fit test determines whether a sample distribution matches an expected distribution.
   - It compares observed frequencies to expected frequencies across categories.
   - If the difference is large, it suggests the data does not fit the expected distribution.

19. What is the F-distribution, and when is it used in hypothesis testing
   - The F-distribution is used to compare two variances and is the basis for the ANOVA test.
   - It's positively skewed and depends on two degrees of freedom: one for the numerator and one for the denominator.
   - It helps assess whether groups have significantly different variances.

20. What is an ANOVA test, and what are its assumptions
   - ANOVA (Analysis of Variance) tests whether there are significant differences between the means of three or more groups.
   - It assumes independence of observations, normally distributed groups, and equal variances.
   - It uses the F-statistic to evaluate the null hypothesis that all group means are equal.

21. What are the different types of ANOVA tests
   - The main types of ANOVA are one-way ANOVA (one independent variable), two-way ANOVA (two independent variables), and repeated measures ANOVA (same subjects tested across conditions).
   - Each type is used based on the experimental design and the number of factors involved.

22. What is the F-test, and how does it relate to hypothesis testing
   - The F-test compares variances to determine if they are significantly different.
   - It is used in ANOVA and regression analysis to assess the overall significance of a model.
   - The F-statistic helps determine whether to reject the null hypothesis that variances or group effects are equal.

```python
# 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
def one_sample_z_test(data, pop_mean, pop_std):
    n = len(data)
    sample_mean = np.mean(data)
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p = 2 * (1 - stats.norm.cdf(abs(z)))
    return z, p

data = np.random.normal(100, 10, 50)
z, p = one_sample_z_test(data, 95, 10)
print("Z:", z, "| P:", p)
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import matplotlib.pyplot as plt

z_stat = 2.1
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(x < -1.96) | (x > 1.96), color='red', alpha=0.3, label="Rejection region")
plt.axvline(z_stat, color='blue', linestyle='--', label=f'Z = {z_stat:.2f}')
plt.legend()
plt.title("Two-Tailed Z-Test Decision Region")
plt.show()
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
def plot_errors(mu0=0, mu1=1, sigma=1, alpha=0.05, n=30):
    x = np.linspace(-4, 4, 1000)
    se = sigma / np.sqrt(n)
    critical = stats.norm.ppf(1 - alpha)
    plt.plot(x, stats.norm.pdf(x, mu0, se), label="Null (Type I)")
    plt.plot(x, stats.norm.pdf(x, mu1, se), label="Alternative (Type II)")
    plt.axvline(critical, color='red', linestyle='--', label=f'Critical Z = {critical:.2f}')
    plt.fill_between(x, stats.norm.pdf(x, mu0, se), where=(x > critical), alpha=0.3, color='red', label='Type I Error')
    plt.fill_between(x, stats.norm.pdf(x, mu1, se), where=(x <= critical), alpha=0.3, color='blue', label='Type II Error')
    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.show()

plot_errors()
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    se = stats.sem(data)
    margin = stats.t.ppf((1 + confidence) / 2., n-1) * se
    return mean - margin, mean + margin

ci = confidence_interval(np.random.normal(100, 15, 50))
print("Confidence Interval:", ci)
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
def margin_of_error(data, confidence=0.95):
    se = stats.sem(data)
    n = len(data)
    moe = stats.t.ppf((1 + confidence) / 2., n-1) * se
    return moe

data = np.random.normal(70, 12, 40)
print("Margin of Error:", margin_of_error(data))
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A):
    numerator = likelihood_B_given_A * prior_A
    denominator = numerator + likelihood_B_given_not_A * prior_not_A
    posterior = numerator / denominator
    return posterior

# Example
P_A = 0.01  # Disease prevalence
P_B_given_A = 0.99  # Test true positive
P_not_A = 0.99
P_B_given_not_A = 0.05  # False positive
print("Posterior Probability (Disease given Positive Test):", bayes_theorem(P_A, P_B_given_A, P_not_A, P_B_given_not_A))
```

```python
# 27. Perform a Chi-square test for goodness of fit with simulated data and analyze results.
import numpy as np
from scipy import stats

observed = np.array([25, 30, 45])
expected_proportions = np.array([1/3, 1/3, 1/3])  # Equal proportions
expected = expected_proportions * observed.sum()  # Scale to match total observed

chi2_stat, p_val = stats.chisquare(f_obs=observed, f_exp=expected)
print("Goodness-of-Fit Chi-square:", chi2_stat, "| P-value:", p_val)
```
